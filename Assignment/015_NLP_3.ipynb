```python
# Read a word from user
word = input("Enter a word: ")
```

```python
# Fetch all meanings
meanings = wordnet.synsets(word)
print("All meanings of the word:")
for meaning in meanings:
    print(f"{meaning.name()}: {meaning.definition()}")
```

```python
# Fetch all noun meanings
noun_meanings = [synset for synset in meanings if synset.pos() == 'n']
print("\nNoun meanings of the word:")
for meaning in noun_meanings:
    print(f"{meaning.name()}: {meaning.definition()}")
```

```python
import spacy
from nltk import pos_tag, word_tokenize
from nltk.chunk import RegexpParser
from nltk.stem import PorterStemmer, LancasterStemmer
from nltk.corpus import wordnet

nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('wordnet')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
# Create a chunk grammar with one regular expression rule:
'''According to the rule you created, your chunk:

    Start with an optional (?) determiner ('DT')
    Can have any number (*) of adjectives (JJ)
    End with a noun  (<NN>)
'''
grammar = "NP: {<DT><JJ><NN>}"
```

```python
# 1: Apply chunking and show the chunks of type <VBD><DT>
words = word_tokenize(text)
tagged = pos_tag(words)    
chunk_grammar = "VBD_DT: {<VBD><DT>}"
chunk_parser = RegexpParser(chunk_grammar)
chunked = chunk_parser.parse(tagged)
chunked
```

```python
# 2: Find all the named entities present in the article
doc = nlp(text)
named_entities =[(ent.text, ent.label_) for ent in doc.ents]
print("Found named entities:",named_entities)
```

```python
# 3: Access all the verbs and apply the Porter Stemmer and Lancaster Stemmer
doc = nlp(text)
verbs = [token.text for token in doc if token.pos_ == 'VERB']
porter = PorterStemmer()
lancaster = LancasterStemmer()
porter_stems = [porter.stem(verb) for verb in verbs]
lancaster_stems = [lancaster.stem(verb) for verb in verbs]

print("Porter Stems :",porter_stems)
print("Lancaster Stems :",lancaster_stems)
```

```python
# 4: Access all the verbs in past tense and lemmatize them
doc = nlp(text)
past_tense_verbs = set(token.lemma_ for token in doc if token.tag_ == 'VBD')

print("Lemmatized past tense verbs:",past_tense_verbs)
```

```python
from itertools import chain


doc = nlp(text)
adjectives = [token.text for token in doc if token.pos_ == 'ADJ']
synonyms = {}
for adj in adjectives:
    synsets = wordnet.synsets(adj, pos=wordnet.ADJ)
    syns = set(chain.from_iterable([syn.lemma_names() for syn in synsets]))
    synonyms[adj] = syns

print("Adjective synonymsb :",synonyms)
```

```python
# 6: Read all the verbs and print their antonyms
doc = nlp(text)
verbs = [token.text for token in doc if token.pos_ == 'VERB']
antonyms = {}
for verb in verbs:
    synsets = wordnet.synsets(verb, pos=wordnet.VERB)
    ants = set()
    for syn in synsets:
        for lemma in syn.lemmas():
            if lemma.antonyms():
                ants.add(lemma.antonyms()[0].name())
    antonyms[verb] = ants

print("Verb antonyms:",antonyms)
```

```python
# 7: Print all the entities marked as DATE and TIME in the data
doc = nlp(text)
dates_times = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in {'DATE', 'TIME'}]

print("Entities marked as DATE and TIME:",dates_times)
```

```python
# 8: Print all the names of locations present in the article
doc = nlp(text)
locations = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ == 'GPE']

print("Names of locations in the article:",locations)
```
